```python
import pickle
import os
from pyvi import ViTokenizer
import string
from tqdm import tqdm
```

```python
file_list = []
for (dirpath, dirname, filename) in os.walk('data/'):
    for f in filename:
        file_list.append(dirpath+'/'+f)
```

```python
len(file_list)
```

```python
def clean_document(doc):
    doc = ViTokenizer.tokenize(doc)
    doc = doc.lower() #Lower
    
    tokens = doc.split() #Split in_to words
    
    table = str.maketrans('', '', string.punctuation.replace("_", "")) #Remove all punctuation
    tokens = [w.translate(table) for w in tokens]
    
    tokens = [word for word in tokens if word]
    
    return tokens
```

```python
INPUT_LENGTH = 50
sequences = []
count = 0
for f in tqdm(file_list):
    count += 1
    
    if count <= 10686*3:
        continue
    
    if count%10686==0:
        with open('sequences' + str(count/10686) + '.pkl', 'wb') as f2:
            pickle.dump(sequences, f2)
        sequences = []
    
    f1 = open(f, encoding='utf-16')
    doc = f1.read()
    tokens = clean_document(doc)
    
    for i in range(INPUT_LENGTH + 1, len(tokens)):
        seq = tokens[i-INPUT_LENGTH-1:i]
        line = ' '.join(seq)
        sequences.append(line)
```

```python
f1 = open(file_list[2], encoding='utf-16')
clean_document(f1.read())
```
